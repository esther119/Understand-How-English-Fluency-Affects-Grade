```python
import numpy as np

# df = pd.DataFrame({'docId': [1,2,3], 
#                'sent': ['This is the first sentence', 
#                         'This is the second sentence',
#                         'This is the third sentence']})

# Tokenize and generate count vectors
word_vec = df.Title.apply(str.split).apply(pd.value_counts).fillna(0)

# Compute term frequencies
tf = word_vec.divide(np.sum(word_vec, axis=1), axis=0)

# Compute inverse document frequencies
idf = np.log10(len(tf) / word_vec[word_vec > 0].count()) 

# Compute TF-IDF vectors
tfidf = np.multiply(tf, idf.to_frame().T)

print(tfidf)
```

```python
tfidf.sum().sort_values(ascending=False)
```

```python
tfidf_words = tfidf.sum().to_frame(name="tf-idf")
tfidf_words
```

```python
tfidf_words.rank()
```
