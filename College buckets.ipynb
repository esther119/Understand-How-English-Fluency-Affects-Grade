```python
import scipy.stats as sts
def correlation_comparison(dataset, y_column): 
    x= np.array(dataset['Assessment reports Score']) 
    y = np.array(dataset[y_column])


    code_coef = np.corrcoef(x, y)
    r_pear, p_pear = sts.pearsonr(x, y)
    r_spear, p_spear = sts.spearmanr(x, y)
    r_kend, p_kend = sts.kendalltau(x, y)

    r_z = np.arctanh(r_pear)

    print(f'Mean of readability:{np.mean(y)} and std of readability:{np.std(y)}')
    plt.title(f"Fluency score distribution in the {dataset.name} group")
    sns.distplot(y)
    # plt.text(0.5, 0.5, 'matplotlib', ha='right', va='top')

    plt.annotate(f'Mean:{round(np.mean(y),2)}, std:{round(np.std(y),2)}', xy=(0.05, 0.95), xycoords='axes fraction')
    plt.show()

    print(f'The correlation between Assessment reports Score and {y_column}')
    print(f'Person r and p value:{r_pear, p_pear}')
    print(f'Spearman r and p value:{r_spear, p_spear}')
    print(f'Kendalltau r and p value:{r_kend, p_kend}')
    print(f'Fisher z-transformation value:{r_z}')
    
    sns.set_palette('rainbow')
    sns.boxplot(x=x, y=y)
    plt.xlabel('Grade')
    plt.ylabel('Fluency score')
    plt.title(f"Fluency score by grade in the {dataset.name} group")
    plt.show()

    sns.violinplot(data=df, x=x, y = y,  scale='width')
    plt.xlabel('Grade')
    plt.ylabel('Fluency score')    
    plt.title(f"Fluency score by grade in the {dataset.name} group")
    plt.show()
```

```python
sns.distplot(CS_score, hist=False, label= 'CS')
sns.distplot(NS_score, hist=False, label= 'NS')
sns.distplot(SS_score, hist=False, label= 'SS')
sns.distplot(AH_score, hist=False, label= 'AH')
plt.title('Fluency score distribution for different colleges')
plt.xlabel('Fluency score')
plt.legend()
```

```python
def comparison_stats(dataset1, dataset2):    
    effect_size = pg.compute_effsize(dataset1, dataset2, paired=False, eftype='cohen')
    n_dataset1 = len(dataset1)
    s_dataset1 = np.var(dataset1)
    n_dataset2 = len(dataset2)
    s_dataset2 = np.var(dataset2)

    std = np.sqrt(((n_dataset1 - 1) * n_dataset2 + (n_dataset2 - 1) * s_dataset2) / (n_dataset1 + n_dataset2 - 2))
    t_test = pg.ttest(dataset1, dataset2,correction=True)
    return [effect_size, std, t_test['p-val'][0], t_test['T'][0]]

table_data = []   
AH_CS = comparison_stats(AH_score,CS_score)
AH_SS = comparison_stats(AH_score,SS_score)
AH_NS = comparison_stats(AH_score,NS_score)
NS_CS = comparison_stats(NS_score,CS_score)
SS_CS = comparison_stats(SS_score,CS_score)
NS_SS = comparison_stats(NS_score,SS_score)

table_data.append(['AH vs CS', AH_CS[0], AH_CS[1], AH_CS[2], AH_CS[3]])
table_data.append(['AH vs SS', AH_SS[0], AH_SS[1], AH_SS[2], AH_SS[3]])
table_data.append(['AH vs NS', AH_NS[0], AH_NS[1], AH_NS[2], AH_NS[3]])
table_data.append(['NS vs CS', NS_CS[0], NS_CS[1], NS_CS[2], NS_CS[3]])
table_data.append(['SS vs CS', SS_CS[0], SS_CS[1], SS_CS[2], SS_CS[3]])
table_data.append(['NS vs SS', NS_SS[0], NS_SS[1], NS_SS[2], NS_SS[3]])


plot_df = pd.DataFrame(table_data, columns=['Comparison', 'Effect_size', 'Error', 'P value', 'T'])
plot_df
```

## Grade scale comparison

```python
def distribution_plot(input_texts, name): 
    i20 = input_texts['dale_chall_readability_score'].quantile(0.2)
    i80 = input_texts['dale_chall_readability_score'].quantile(0.8)

    small = input_texts[input_texts['dale_chall_readability_score'].lt(i20)]
    great = input_texts[input_texts['dale_chall_readability_score'].gt(i80)]

    x = small['Assessment reports Score']
    y = great['Assessment reports Score']
    print("group size", x.shape, y.shape)

    bins1 = [0.3,1.3,2.3,3.3,4.3,5.3]
    bins2 = [-0.7,0.7,1.7,2.7,3.7,4.7]
    bin_counts, _, _ = plt.hist(x, color='r', alpha=0.5, bins = bins1, rwidth= 0.4, density = True, label = 'Low Fluency')
    bin_percentages = bin_counts / bin_counts.sum() * 100
    y_ticks = np.arange(0, 70 + 1, 10)
    plt.yticks(y_ticks / 100, y_ticks)
    plt.ylim(0, 0.6)  # Set y-tick range to 0-70%


    plt.hist(y, color='b', alpha=0.5, bins = bins2, rwidth = 0.4, density = True, label = 'High Fluency')
    plt.legend(loc='upper right',prop={'size': 7})
    plt.grid(axis='y', linewidth=0.5, linestyle='dashed', color='lightgray')
    plt.ylabel('Percentage of polls')
    plt.xlabel('Grade')
    plt.title(f'Percentage of polls by grade for {name} data')

    small2 = sum([small['Assessment reports Score'] == 2]).sum() / len(small) * 100
    small3 = sum([small['Assessment reports Score'] == 3]).sum() / len(small) * 100
    small4 = sum([small['Assessment reports Score'] == 4]).sum() / len(small) * 100

    great2 = sum([great['Assessment reports Score'] == 2]).sum() / len(great) * 100
    great3 = sum([great['Assessment reports Score'] == 3]).sum() / len(great) * 100
    great4 = sum([great['Assessment reports Score'] == 4]).sum() / len(great) * 100
    
    print("Percentage of grade 2 for bottom vs top 20 percent fluency: ", small2.round(2), ' vs ', great2.round(2))
    print("Percentage of grade 3 for bottom vs top 20 percent fluency: ", small3.round(2),' vs ', great3.round(2))
    print("Percentage of grade 4 for bottom vs top 20 percent fluency: ", small4.round(2),' vs ', great4.round(2))

    print(f'Average, std grade for top 30 percent fluency: {great["Assessment reports Score"].mean()}, {great["Assessment reports Score"].std()}')
    print(f'Average, std grade for bottom 30 percent fluency: {small["Assessment reports Score"].mean()}, {small["Assessment reports Score"].std()}')
    
    
    print("Mean diff:", great['Assessment reports Score'].mean() - small['Assessment reports Score'].mean())
    print("One std diff:", great['Assessment reports Score'].mean() + great['Assessment reports Score'].std() - (small['Assessment reports Score'].mean() + small['Assessment reports Score'].std()))

distribution_plot(AH_data, 'AH')
```
